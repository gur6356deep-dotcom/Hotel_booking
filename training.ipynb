```python
num = []
cat = []

for x in data.columns:
    if data[x].dtype == "object":
        cat.append(x)
    else:
        num.append(x)

print("Numerical Columns:", num)
print("Categorical Columns:", cat)
```

```python
for col in num:
    plt.figure(figsize=(8,5))
    plt.hist(data[col], bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
```

```python
for x in data.columns:

    print(x)

    if data[x].dtype == "object":
        print(data[x].value_counts())

    else:
        print(data[x].describe())

    print("-"*50)
```

```python
plt.figure(figsize=(8,8))

data['hotel'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=90,
    shadow=True
)

plt.title("Distribution of Hotel Types")
plt.ylabel("")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='hotel', data=data)
plt.title("Bar Plot of Hotel Type")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(data['adr'], bins=30)
plt.title("Histogram of ADR")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(data['lead_time'], data['adr'])
plt.title("Scatter Plot of Lead Time vs ADR")
plt.xlabel("Lead Time")
plt.ylabel("ADR")
plt.show()
```

```python
plt.figure(figsize=(10,5))
data['arrival_date_month'].value_counts().sort_index().plot()
plt.title("Line Plot of Bookings by Month")
plt.show()
```
